```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data
angle = [40.58, 34.95, 30.73, 32.93, 40.94, 60.44, 84.98]
temp = [4012, 4013, 4012.1, 4014.25, 4012, 4011, 4005.3]

y_err = [1, 1, 1, 1, 1, 1, 1]

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(angle, temp)
angles=range(0,90)
line = [slope * x + intercept for x in angles]
print(intercept)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(angle, temp)
plt.errorbar(angle, temp, yerr=y_err, fmt='o', capsize=5)
plt.plot(angles, line, color='red', label=f'Best Fit Line\n$R^2$ = {r_value**2:.4f}')
plt.xlim(0, 90)
plt.ylim(4000, 4100)
plt.xlabel(r'Zenith Angle (Degrees)')
plt.ylabel(r'Temperature (K)')
plt.title('Plot of Temperatures with their Respective Zenith Angles')
plt.legend()
plt.grid(True, which="both", linestyle="--", alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.62607015e-34  # Planck's constant (J·s)
c = 2.99792458e8    # Speed of light (m/s)
k = 1.380649e-23    # Boltzmann's constant (J/K)

# Planck's Law
def planck(wavelength, T):
    """
    Calculate spectral radiance using Planck's Law.
    Wavelength in meters, temperature in Kelvin.
    """
    a = 2.0 * h * c**2
    b = h * c / (wavelength * k * T)
    return a / (wavelength**5 * (np.exp(b) - 1.0))

# Wavelength range: 0.1 to 3 microns (UV to infrared)
wavelengths = np.linspace(0.1e-6, 3e-6, 500)  # meters
temperature = 5778  # Sun's surface temperature in K
intensity = planck(wavelengths, temperature)
print(planck(500e-9,5778),planck(655e-9,5778))
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(wavelengths * 1e9, intensity, color='orange')  # Convert x-axis to nm
plt.axvline(x=655, color='r', linestyle='--', linewidth=2, label='Effective Observed Wavelength (655nm)')
plt.legend()
plt.title("Blackbody Radiation Curve for the Sun (T = 5778 K)")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Spectral Radiance (W·sr⁻¹·nm⁻³)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(100, 3000)
plt.ylim(bottom=0)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.62607015e-34  # Planck's constant (J·s)
c = 2.99792458e8    # Speed of light (m/s)
k = 1.380649e-23    # Boltzmann's constant (J/K)

# Planck's Law
def planck(wavelength, T):
    a = 2.0 * h * c**2
    b = h * c / (wavelength * k * T)
    return a / (wavelength**5 * (np.exp(b) - 1.0))

# Wavelength range: 0.1 to 3 microns (UV to infrared)
wavelengths = np.linspace(0.1e-6, 3e-6, 500)  # meters

# Temperatures
temps = [5778, 4366]

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

for i, T in enumerate(temps):
    intensity = planck(wavelengths, T)
    axs[i].plot(wavelengths * 1e9, intensity, color='orange' if T == 5778 else 'darkred')
    axs[i].axvline(x=655, color='r', linestyle='--', linewidth=2, label='Observed Wavelength (655 nm)')
    axs[i].set_title(f"Blackbody Curve (T = {T} K)")
    axs[i].set_xlabel("Wavelength (nm)")
    axs[i].set_xlim(100, 3000)
    axs[i].grid(True, linestyle='--', alpha=0.6)
    axs[i].legend()

axs[0].set_ylabel("Spectral Radiance (W·sr⁻¹·nm⁻³)")
plt.suptitle("Blackbody Radiation Comparison: Sun vs Cooler Star")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

print(f"The peak wavelength of the second blackbody is {2.9e-3/4366}")

print(planck(500e-9,5778)/planck(665e-9,4366))
print()
```
